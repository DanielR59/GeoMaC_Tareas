```python
#importamos librerias
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
y1 = lambda x: 1-np.exp(1)**x+(np.exp(1)-2)*np.sin(np.pi*x/2)
dy1= lambda x: -(np.exp(1)**x)+(np.pi*(np.exp(1)-1)/2)*np.cos(np.pi*x/2)

xmin = 0
xmax = 1.1
x = np.linspace(xmin, xmax, 100)

ymin = np.min(y1(x))
ymax = np.max(y1(x))

A=0; B=1 #intervalos del 0

plt.plot(x,y1(x),label='Funcion original')
plt.plot(x,dy1(x),label='derivada')
plt.plot([A,A], [ymin,ymax], 'g--', lw=1)
plt.plot([B,B], [ymin,ymax], 'g--', lw=1)
plt.plot([xmin, xmax], [0, 0])
plt.legend()
plt.show()
```

```python
def Errores(ValorReal,ValorAprox):
    Ea=np.fabs(ValorReal-ValorAprox); Er=Ea/np.fabs(ValorReal)
    print('Error absoluto ={}'.format(Ea))
    print('Error relativo ={}'.format(Er))
    #return Ea,Er
    
p=np.pi; pa=22/7
Errores(p,pa)

p=np.exp(10); pa=22000
Errores(p,pa)
p=10**(np.pi); pa=1400
Errores(p,pa)
p=math.factorial(9); pa=np.sqrt(18*np.pi)*(9/np.exp(1))**9
Errores(p,pa)
```
